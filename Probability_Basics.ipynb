1. A die is rolled. What is the probability of getting:
(a) An even number
(b) A number greater than 4

-   When a fair die is rolled, the sample space is:
{1, 2, 3, 4, 5, 6} → total outcomes = 6

(a) Probability of getting an even number

Even numbers on a die: {2, 4, 6} → 3 outcomes

Probability = Favorable outcomes / Total outcomes
= 3 / 6 = 1/2

(b) Probability of getting a number greater than 4

Numbers greater than 4: {5, 6} → 2 outcomes

Probability = 2 / 6 = 1/3

 Final Answers:
(a) 1/2
(b) 1/3

4. The population of a school is divided into 60% boys and 40% girls. If you want equal representation
of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?
Why?
-    Correct Method: Stratified Sampling
 Why not Simple Random Sampling?

In Simple Random Sampling, every student has an equal chance, but:

Since the population is 60% boys and 40% girls, your sample will likely reflect the same imbalance.
You cannot guarantee equal representation of boys and girls.
 Why Stratified Sampling is better:

In Stratified Sampling:

You divide the population into groups (called strata) → Boys and Girls
Then you select equal numbers from each group.

 This ensures:

Balanced representation (50% boys, 50% girls)
More accurate comparison between genders
 Final Answer:

Use Stratified Sampling, because it allows you to intentionally select equal numbers from both boys and girls, ensuring fair representation.

5. The average height of 1000 students = 160 cm. A sample of 100 students shows an average height
= 158 cm.
Find the sampling error

-    Sampling Error Formula:

Sampling Error = Sample Mean − Population Mean

Given:
Population mean = 160 cm
Sample mean = 158 cm
Calculation:

Sampling Error = 158 − 160 = −2 cm

 Final Answer:

Sampling Error = −2 cm

 Insight:
The negative sign means the sample underestimates the population average.
The sample’s average height is 2 cm less than the actual population average.

10. Differentiate between discrete and continuous random variables with examples

-    Discrete vs Continuous Random Variables
| Feature        | Discrete Random Variable                    | Continuous Random Variable                             |
| -------------- | ------------------------------------------- | ------------------------------------------------------ |
| **Definition** | Takes **countable values**                  | Takes **uncountable (infinite) values** within a range |
| **Values**     | Specific, separate numbers (0,1,2,3…)       | Any value in an interval (e.g., 1.2, 1.25, 1.257…)     |
| **Nature**     | Countable (you can list them)               | Measurable (cannot list all values)                    |
| **Examples**   | Number of students in class, number of cars | Height, weight, temperature                            |
| **Graph Type** | Bar graph                                   | Smooth curve (like bell curve)                         |


Discrete Random Variable (Example)
Number of heads when tossing 3 coins → {0, 1, 2, 3}
Number of customers in a shop

 Values are fixed and countable

 Continuous Random Variable (Example)
Height of students (e.g., 160.2 cm, 160.25 cm…)
Time taken to complete a task

 Values can be any number within a range

 Key Difference (Simple Trick):
Discrete = Count (गणना)
Continuous = Measure (मापन)
 Final Summary:
Discrete variables → Countable values (integers)
Continuous variables → Infinite measurable values
